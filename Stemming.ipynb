```python
import nltk
import pandas as pd
import ast
from nltk.stem import RSLPStemmer

nltk.download("rslp")

df = pd.read_csv("todas_as_praias_remove_stopwords.csv", encoding="utf-8-sig")

stemmer = RSLPStemmer()

def converter_para_lista(text):
    if not isinstance(text, (str, list)):
        return []

    if isinstance(text, str):
        try:
            text = ast.literal_eval(text)
            if not isinstance(text, list):
                text = [text]
        except (ValueError, SyntaxError):
            text = text.split()

    return text

def aplicar_stemming(tokens):
    tokens = converter_para_lista(tokens)
    return [stemmer.stem(token) for token in tokens if isinstance(token, str) and token.strip()]

df["title_stemming"] = df["title_remove_stopwords"].apply(aplicar_stemming)
df["text_stemming"]  = df["text_remove_stopwords"].apply(aplicar_stemming)

df.to_csv("todas_as_praias_remove_stopwords.csv", index=False, encoding="utf-8-sig")
```
