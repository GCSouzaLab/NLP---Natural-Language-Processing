```python
import nltk
import pandas as pd
import ast
from nltk.corpus import stopwords

df = pd.read_csv("todas_as_praias_tokenizado.csv", encoding="utf-8-sig")

def remover_stopwords(text):
    # Retorna lista vazia para qualquer valor que não seja string ou lista
    if not isinstance(text, (str, list)):
        return []

    # Converte string "['a', 'b']" de volta para lista, se necessário
    if isinstance(text, str):
        try:
            text = ast.literal_eval(text)
            # Se o resultado não for lista, trata como texto simples
            if not isinstance(text, list):
                text = [text]
        except (ValueError, SyntaxError):
            text = text.split()

    stop_words = set(stopwords.words('portuguese'))
    return [w for w in text if w not in stop_words]

df["title_remove_stopwords"] = df["title_tokens"].apply(remover_stopwords)
df["text_remove_stopwords"]  = df["text_tokens"].apply(remover_stopwords)

df.to_csv("todas_as_praias_remove_stopwords.csv", index=False, encoding="utf-8-sig")

print(df[["title", "title_remove_stopwords", "text", "text_remove_stopwords"]].head())
```
