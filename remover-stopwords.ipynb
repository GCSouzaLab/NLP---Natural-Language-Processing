```python
import nltk
import pandas as pd
import ast
import unicodedata
from nltk.corpus import stopwords

# Caso ainda não tenha baixado
nltk.download("stopwords")

df = pd.read_csv("todas_as_praias_tokenizado.csv", encoding="utf-8-sig")

def converter_para_lista(text):
    # Retorna lista vazia para qualquer valor que não seja string ou lista
    if not isinstance(text, (str, list)):
        return []

    # Se vier como string do tipo "['a', 'b']", converte para lista
    if isinstance(text, str):
        try:
            text = ast.literal_eval(text)
            if not isinstance(text, list):
                text = [text]
        except (ValueError, SyntaxError):
            text = text.split()

    return text

def normalizar_tokens(text):
    tokens = converter_para_lista(text)

    tokens_normalizados = []
    for w in tokens:
        if not isinstance(w, str):
            continue

        # minúsculas + remoção de acentos
        w = w.lower()
        w = unicodedata.normalize("NFD", w)
        w = "".join(c for c in w if unicodedata.category(c) != "Mn")

        tokens_normalizados.append(w)

    return tokens_normalizados

def remover_stopwords(text):
    tokens = converter_para_lista(text)
    stop_words = {'a', 'ao', 'aos', 'aquela', 'aquelas', 'aquele', 'aqueles', 'aquilo', 'as', 'até', 'com', 'como', 'da', 'das', 'de', 'dela', 'delas', 'dele', 'deles', 'depois', 'do', 'dos', 'e', 'ela', 'elas', 'ele', 'eles', 'em', 'entre', 'era', 'eram', 'essa', 'essas', 'esse', 'esses', 'esta', 'estamos', 'estar', 'estas', 'estava', 'estavam', 'este', 'esteja', 'estejam', 'estejamos', 'estes', 'esteve', 'estive', 'estivemos', 'estiver', 'estivera', 'estiveram', 'estiverem', 'estivermos', 'estivesse', 'estivessem', 'estivéramos', 'estivéssemos', 'estou', 'está', 'estávamos', 'estão', 'eu', 'foi', 'fomos', 'for', 'fora', 'foram', 'forem', 'formos', 'fosse', 'fossem', 'fui', 'fôramos', 'fôssemos', 'haja', 'hajam', 'hajamos', 'havemos', 'haver', 'hei', 'houve', 'houvemos', 'houver', 'houvera', 'houveram', 'houverei', 'houverem', 'houveremos', 'houveria', 'houveriam', 'houvermos', 'houverá', 'houverão', 'houveríamos', 'houvesse', 'houvessem', 'houvéramos', 'houvéssemos', 'há', 'hão', 'isso', 'isto', 'já', 'lhe', 'lhes', 'mais', 'mas', 'me', 'mesmo', 'meu', 'meus', 'minha', 'minhas', 'muito', 'na', 'nas', 'nem', 'no', 'nos', 'nossa', 'nossas', 'nosso', 'nossos', 'num', 'numa', 'não', 'nós', 'o', 'os', 'ou', 'para', 'pela', 'pelas', 'pelo', 'pelos', 'por', 'qual', 'quando', 'que', 'quem', 'se', 'seja', 'sejam', 'sejamos', 'sem', 'ser', 'serei', 'seremos', 'seria', 'seriam', 'será', 'serão', 'seríamos', 'seu', 'seus', 'somos', 'sou', 'sua', 'suas', 'só', 'também', 'te', 'tem', 'temos', 'tenha', 'tenham', 'tenhamos', 'tenho', 'terei', 'teremos', 'teria', 'teriam', 'terá', 'terão', 'teríamos', 'teu', 'teus', 'teve', 'tinha', 'tinham', 'tive', 'tivemos', 'tiver', 'tivera', 'tiveram', 'tiverem', 'tivermos', 'tivesse', 'tivessem', 'tivéramos', 'tivéssemos', 'tu', 'tua', 'tuas', 'tém', 'tínhamos', 'um', 'uma', 'você', 'vocês', 'vos', 'à', 'às', 'é', 'éramos'}
    return [w for w in tokens if w not in stop_words]

# 1) cria colunas normalizadas
df["title_tokens_normalized"] = df["title_tokens"].apply(normalizar_tokens)
df["text_tokens_normalized"]  = df["text_tokens"].apply(normalizar_tokens)

# 2) remove stopwords a partir das colunas normalizadas
df["title_remove_stopwords"] = df["title_tokens_normalized"].apply(remover_stopwords)
df["text_remove_stopwords"]  = df["text_tokens_normalized"].apply(remover_stopwords)

df.to_csv("todas_as_praias_remove_stopwords.csv", index=False, encoding="utf-8-sig")
```
