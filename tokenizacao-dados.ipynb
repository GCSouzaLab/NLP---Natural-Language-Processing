Armazena todos os dados em um único .CSV, coloca o nome do arquivo (praia) como coluna no novo .CSV

```python
import csv

import pandas as pd
import glob
import os

COLUNAS = ["rating", "title", "text", "location", "num_contributions", "date"]

arquivos = glob.glob("csv/*.csv")

dfs = []
for arquivo in arquivos:
    try:

        delimiter = ','

        if arquivo == "csv\praia_bacia_da_vovó.csv" or arquivo == "csv\praia_do_poá.csv":
            delimiter = ';'

        df = pd.read_csv(arquivo, encoding="utf-8", engine="python", quotechar='"', delimiter=delimiter)
        df["praia"] = os.path.splitext(os.path.basename(arquivo))[0]
        dfs.append(df)
        print(f"✅ {arquivo} — {len(df)} linhas")

    except Exception:
        # Tenta ler linha a linha, ignorando as linhas problemáticas
        try:
            df = pd.read_csv(arquivo, encoding="utf-8", engine="python",
                             quotechar='"', on_bad_lines="skip", names=COLUNAS,
                             header=0)
            df["praia"] = os.path.splitext(os.path.basename(arquivo))[0]
            dfs.append(df)
            print(f"⚠️  {arquivo} — {len(df)} linhas (algumas ignoradas)")

        except Exception as e:
            print(f"❌ Não foi possível ler {arquivo}: {e}")

df_final = pd.concat(dfs, ignore_index=True)
df_final.to_csv("todas_as_praias.csv", index=False, encoding="utf-8-sig")

print(f"\n📊 Total: {len(df_final)} reviews de {len(dfs)} arquivos")
print(df_final["praia"].value_counts())
```

```python
import pandas as pd
import nltk
import string

nltk.download('punkt')
nltk.download('punkt_tab')

df = pd.read_csv("todas_as_praias.csv", encoding="utf-8-sig")

def tokenizar(texto):
    tokens = nltk.word_tokenize(str(texto))
    # Remove pontuação
    tokens = [t for t in tokens if t not in string.punctuation]
    # Junta como string separada por espaço
    return " ".join(tokens)

df["title_tokens"] = df["title"].apply(tokenizar)
df["text_tokens"]  = df["text"].apply(tokenizar)

df.to_csv("todas_as_praias_tokenizado.csv", index=False, encoding="utf-8-sig")

print(df[["title", "title_tokens", "text", "text_tokens"]].head())
```
