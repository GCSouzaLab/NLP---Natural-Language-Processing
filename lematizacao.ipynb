```python
import csv

import spacy
import pandas as pd

df = pd.read_csv("todas_as_praias_remove_stopwords.csv", encoding="utf-8-sig")

nlp = spacy.load("pt_core_news_lg")

input_csv_path = 'input_texts.csv'
df.to_csv(input_csv_path, index=False)
print(f"Created input file: {input_csv_path}")


def lemmatize_text(text):
    doc = nlp(text)
    lemmas = [token.lemma_ for token in doc]
    return ' '.join(lemmas)


df = pd.read_csv(input_csv_path)

df["title_lemmatized"] = df["title_remove_stopwords"].apply(lemmatize_text)
df["text_lemmatized"]  = df["text_remove_stopwords"].apply(lemmatize_text)


df.to_csv("todas_as_praias_lemmatized.csv", index=False, encoding="utf-8-sig")

print("\nResults:")
print(df)
```
